```python
import matplotlib.pylab as plt
import time
import keras
from keras.datasets import mnist,cifar10
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU, Conv2D,MaxPool2D,Flatten,Dropout
from keras.optimizers import RMSprop
from keras import backend as K
K.tensorflow_backend._get_available_gpus()
from matplotlib.pyplot import imshow
import numpy as np

import tensorflow as tf
gpu_options = tf.GPUOptions(allow_growth=True)
sess = tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
tf.keras.backend.set_session(sess)
```

# Load Data 

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape((x_train.shape[0],x_train.shape[1]*x_train.shape[1]))
x_test = x_test.reshape((x_test.shape[0],x_test.shape[1]*x_test.shape[1]))
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

# Part1 

### Training 

```python
batch_sizes = [64,1024]
historys = []
models_weights = []
for b in batch_sizes:
    model = Sequential()
    model.add(Dense(64,kernel_initializer='random_normal', activation='relu', input_dim=784))
    model.add(Dense(128,kernel_initializer='random_normal', activation='relu'))
    model.add(Dense(64,kernel_initializer='random_normal', activation='relu'))
    model.add(Dense(10,kernel_initializer='random_normal', activation='softmax'))

    model.compile(loss='categorical_crossentropy',optimizer= 'adam',metrics=['accuracy'])
    batch_size = b
    epochs = 500
    history = model.fit(x_train, y_train,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(x_test, y_test))
    historys.append(history)
    models_weights.append(model.get_weights())
```

```python
def get_interpolation_parameters(p1,p2):
    alpha = np.linspace(-1,2,100)
    p_list = []
    for a in alpha:
        p_list.append(p1*(1-a) + p2*a)
    return alpha,p_list
```

```python
model = Sequential()
model.add(Dense(64,activation='relu', input_dim=784))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(10,activation='softmax'))

loss_l = []
acc_l = []
val_loss_l = []
val_acc_l = []

alpha,p_list = get_interpolation_parameters(np.array(models_weights[0]),np.array(models_weights[1]))

for weight in p_list: 
    model.set_weights(weight)
    model.compile(loss='categorical_crossentropy',optimizer= 'adam',metrics=['accuracy'])
    
    loss, acc = model.evaluate(x_train, y_train)
    val_loss, val_acc = model.evaluate(x_test, y_test)

    loss_l.append(loss)
    acc_l.append(acc)
    val_loss_l.append(val_loss)
    val_acc_l.append(val_acc)
```

```python
fig = plt.figure(figsize = (15,8))
plt.title('Mnist different batch size')

# loss graph
loss_graph = fig.add_subplot(111)
train_loss = loss_graph.plot(alpha,loss_l,label = 'train',color = 'b')
test_loss = loss_graph.plot(alpha,val_loss_l,label = 'test',color = 'b',linestyle =  '--')
plt.ylabel('cross_entrophy')

# accuracy graph
acc_graph = fig.add_subplot(111,sharex = loss_graph,frameon = False)
traina_acc = acc_graph.plot(alpha,acc_l,label = 'train',color = 'r')
test_acc = acc_graph.plot(alpha,val_acc_l,label = 'test',color = 'r',linestyle = '--')
plt.xlabel('alpha')
acc_graph.yaxis.tick_right() # make ytick to the right
acc_graph.yaxis.set_label_position("right") #make ylabel to the right
plt.ylabel('accuracy')
plt.legend(loc = 'best')

plt.show()
```

# Part2 

```python
# get the gradients for the loss to the input
def get_sensitivity(model,data):
    gradients = K.gradients(model.output,model.input)

    sess = tf.InteractiveSession()
    sess.run(tf.initialize_all_variables())
    evaluated_gradients = sess.run(gradients,feed_dict={model.input:data})
    sess.close()
    sensitivity = np.sqrt(np.sum(np.array(evaluated_gradients)**2))
    
    return sensitivity
    
```

```python
batch_sizes = [8,10,16,64,100,128,512,1024,2048,4096]

loss_l = []
acc_l = []
val_loss_l = []
val_acc_l = []

sensitivitys = []
for b in batch_sizes:
    model = Sequential()
    model.add(Dense(32,kernel_initializer='random_normal', activation='relu', input_dim=784))
    model.add(Dense(64,kernel_initializer='random_normal', activation='relu'))
    model.add(Dense(32,kernel_initializer='random_normal', activation='relu'))
    model.add(Dense(10,kernel_initializer='random_normal', activation='softmax'))

    model.compile(loss='categorical_crossentropy',optimizer= 'adam',metrics=['accuracy'])
    batch_size = b
    epochs = 50
    history = model.fit(x_train, y_train,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(x_test, y_test))
    
    loss_l.append(history.history['loss'][-1])
    val_loss_l.append(history.history['val_loss'][-1])
    acc_l.append(history.history['acc'][-1])
    val_acc_l.append(history.history['val_acc'][-1])
    
    sensitivitys.append(get_sensitivity(model,x_train))
    
```

```python
fig = plt.figure(figsize = (15,8))
plt.title('Mnist different batch size')

loss_graph = fig.add_subplot(111)
train_loss = loss_graph.plot(batch_sizes,loss_l,label = 'train',color = 'b')
test_loss = loss_graph.plot(batch_sizes,val_loss_l,label = 'test',color = 'b',linestyle = '--')
plt.ylabel('cross_entrophy')
plt.legend(loc = 'upper left')

s_graph = fig.add_subplot(111,sharex = loss_graph,frameon = False)
traina_s = s_graph.plot(batch_sizes,sensitivitys,label = 'sensitivity',color = 'r')
plt.xlabel('batch sizes')
s_graph.yaxis.tick_right() # make ytick to the right
s_graph.yaxis.set_label_position("right") #make ylabel to the right
plt.ylabel('sensitivity')
plt.legend(loc = 'upper right')

plt.show()
```

```python
fig = plt.figure(figsize = (15,8))
plt.title('Mnist different batch size')

acc_graph = fig.add_subplot(111)
train_acc = acc_graph.plot(batch_sizes,acc_l,label = 'train',color = 'b')
test_acc = acc_graph.plot(batch_sizes,val_acc_l,label = 'test',color = 'b',linestyle = '--')
plt.ylabel('cross_entrophy')
plt.legend(loc = 'upper left')

s_graph = fig.add_subplot(111,sharex = loss_graph,frameon = False)
traina_s = s_graph.plot(batch_sizes,sensitivitys,label = 'sensitivity',color = 'r')
plt.xlabel('batch sizes')
s_graph.yaxis.tick_right() # make ytick to the right
s_graph.yaxis.set_label_position("right") #make ylabel to the right
plt.ylabel('sensitivity')
plt.legend(loc = 'upper right')

plt.show()
```
