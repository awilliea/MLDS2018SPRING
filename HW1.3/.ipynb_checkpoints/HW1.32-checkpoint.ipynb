```python
import matplotlib.pylab as plt
import time
import keras
from keras.datasets import mnist,cifar10
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU, Conv2D,MaxPool2D,Flatten,Dropout
from keras.optimizers import RMSprop
from keras import backend as K
K.tensorflow_backend._get_available_gpus()
from matplotlib.pyplot import imshow
import numpy as np

import tensorflow as tf
gpu_options = tf.GPUOptions(allow_growth=True)
sess = tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
tf.keras.backend.set_session(sess)
```

# Load Data 

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape((x_train.shape[0],x_train.shape[1]*x_train.shape[1]))
x_test = x_test.reshape((x_test.shape[0],x_test.shape[1]*x_test.shape[1]))
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

# Training

```python
historys = []
models_parameters = []

first_layer_parameters = [1,4,8,16,32,64]
second_layer_parameters = [1,4,8,16,32,64]
third_layer_parameters = [1,8,32]

for f in first_layer_parameters:
    for s in second_layer_parameters:
        for t in third_layer_parameters:
            model = Sequential()
            model.add(Dense(f,kernel_initializer='random_normal', activation='relu', input_dim=784))
            model.add(Dense(s,kernel_initializer='random_normal', activation='relu'))
            model.add(Dense(t,kernel_initializer='random_normal', activation='relu'))
            model.add(Dense(10,kernel_initializer='random_normal', activation='softmax'))

            models_parameters.append(model.count_params())

            model.compile(loss='categorical_crossentropy',optimizer= 'adam',metrics=['accuracy'])
            batch_size = 64
            epochs = 50
            history = model.fit(x_train, y_train,batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(x_test, y_test))
            historys.append(history)
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.title('Model Loss')
for i,h in enumerate(historys):
    if(i == 0):
        plt.scatter(models_parameters[i],h.history['loss'][-1],color = 'b',label = 'train_loss')
        plt.scatter(models_parameters[i],h.history['val_loss'][-1],color = 'orange',label = 'test_loss')
    else:
        plt.scatter(models_parameters[i],h.history['loss'][-1],color = 'b')
        plt.scatter(models_parameters[i],h.history['val_loss'][-1],color = 'orange')
plt.legend(loc = 'upper right')
plt.xlabel('number of parameters')
plt.ylabel('loss')

plt.subplot(1,2,2)
plt.title('Model Accuracy')
for i,h in enumerate(historys):
    if(i == 0):
        plt.scatter(models_parameters[i],h.history['acc'][-1],color = 'b',label = 'train_acc')
        plt.scatter(models_parameters[i],h.history['val_acc'][-1],color = 'orange',label = 'test_acc')
    else:
        plt.scatter(models_parameters[i],h.history['acc'][-1],color = 'b')
        plt.scatter(models_parameters[i],h.history['val_acc'][-1],color = 'orange')
plt.legend(loc = 'lower right')
plt.xlabel('number_of_parameters')
plt.ylabel('accuracy')
```
